```python
import csv 
guests = {}
rooms = {}
bookings = {}
food_menu = {
    "1": {"name": "Jollof Rice and Chicken", "price": 2500},
    "2": {"name": "Jollof Rice and Chicken and Fish", "price": 3500},
    "3": {"name": "Rice and Beans With Meat", "price": 2500},
    "4": {"name": "Amala", "price": 500},
    "5": {"name": "Semo", "price": 600},
    "6": {"name": "Pounded Yam", "price": 700},
    "7": {"name": "Suya", "price": 1000},
    "8": {"name": "Shawarma", "price": 2500}
}
amenities = {
    "1": {"name": "Swimming Pool", "price": 2000},
    "2": {"name": "Gym", "price": 1000},
    "3": {"name": "Game House", "price": 2000},
    "4": {"name": "Club House", "price": 2000}
}
drinks_items = {
    "1": {"name": "Pepsi", "price": 500},
    "2": {"name": "Fanta", "price": 500},
    "3": {"name": "Sprite", "price": 500},
    "4": {"name": "Maltina", "price": 700},
    "5": {"name": "Star Lager Beer", "price": 1500},
    "6": {"name": "Guinness Lager Beer", "price": 1500},
    "7": {"name": "Trophy Lager Beer", "price": 1500},
    "8": {"name": "Hero Lager Beer", "price": 1500},
    "9": {"name": "Jameson Whiskey", "price": 1500},
    "10": {"name": "Hennessy Cognac", "price": 1500},
    "11": {"name": "Smirnoff Vodka", "price": 1500}
}
cart = {}
regular_room_price = 5000
special_room_price = 10000
accounts = {}

def creat_bank_account():
    print("Creat a new bank account:")
    while True:
        name = input(f"Enter your name: ")
        if name.replace(" ", "").isalpha():
            break
        else:
            print("Invalid name. Please try again.")
    while True:
        email = input("Enter your email: ")
        if "@" in email and email.endswith(".com"):
            break
        else:
            print("Invalid email. Please enter a correct email.")
    while True:
        phone_number = input("Enter your phone number: ")
        if phone_number.isdigit() and len(phone_number) == 11:
            break
        else:
            print("Invalid phone number. Please enter a correct phone number.")
    while True:
        BVN = input("Enter your BVN: ")
        if BVN.isdigit() and len(BVN) == 11:
            break
        else:
            print("Invalid BVN.Please enter a correct an 11- digit BVN.")
    while True:
        password = input("Enter your password: ")
        if len(password) >= 6:
            break
        else:
            print("Enter too short.Please enter a password with at least 6 characters.")
    while True:
        start_deposit = float(input("Enter initial deposit amount: "))
        if start_deposit >= 0:
            break
        else:
            print("initial deposit must not be nagative number.")
    accounts[name] = {
        "email":email,
        "phone_number":phone_number,
        "password": password,
        "BVN":BVN,
        "balance":start_deposit
    }
    print("Bank account created successfuly!")
def deposit_money():
    name = input("Enter your name: ")
    if name in accounts:
        amount = float(input("Enter amount to deposit: "))
        accounts[name]['balance'] += amount
        print(f"Successfully deposited ₦{amount}. New balance: ₦{accounts[name]['balance']}")
    else:
        print("Account not found.")
def check_balance():
    name = input("Enter your name: ")
    password = input("Enter your password: ")
    if name in accounts and accounts[name]["password"] == password:
        print(f"Current balance: ₦{accounts[name]['balance']}.")
    else:
        print("Invalid email or password.")
def withdraw_money(name, amount):
    if name in accounts:
        if accounts[name]["balance"] >= amount:
            accounts[name]["balance"] -= amount
            return True
        else:
            print("Insufficient funds.")
            return False
    else:
        print("Account not found.")
        return False

            
def add_to_cart(item_type, item_id, guest_name, quantity=1, booking_date="", booking_time=""):
    if item_type == "food":
        item_dict = food_menu
    elif item_type == "amenity":
        item_dict = amenities
    elif item_type == "drink":
        item_dict = drinks_items
    elif item_type == "room":
        item_dict = rooms

    if item_type == "room":
        item_details = {"name": f"Room {item_id}", "price": item_dict[item_id]['price'], "quantity": quantity, "booking_date": booking_date, "booking_time": booking_time}
    else:
        item_details = {"name": item_dict[item_id]['name'], "price": item_dict[item_id]['price'], "quantity": quantity, "booking_date": booking_date, "booking_time": booking_time}
    
    if guest_name not in cart:
        cart[guest_name] = {item_type: [item_details]}
    elif item_type not in cart[guest_name]:
        cart[guest_name][item_type] = [item_details]
    else:
        cart[guest_name][item_type].append(item_details)

def display_cart(guest_name):
    if guest_name in cart:
        print(f"Cart for {guest_name}:")
        for item_type, items in cart[guest_name].items():
            for item in items:
                print(f"{item['name']} x {item['quantity']} - ₦{item['price'] * item['quantity']}")
    else:
        print("Cart is empty.")
print("""
                     **********************************************************************************************
                     *                              Welcome To LEMASK Hotel Management                            *
                     **********************************************************************************************

                       ************************************************************************************
                       *             We offer the best accomodation and services                          *
                       ************************************************************************************

                       *************************************************************************************
                       *                 Please choose an option  from the menu below:                     *
                       *************************************************************************************


""")
while True:
    print("1. Create account")
    print("2. Deposit money")
    print("3. Add guest")
    print("4. Add room")
    print("5. Book room")
    print("6. Check-in")
    print("7. Check-out")
    print("8. Add food menu")
    print("9. Add amenities")
    print("10. Add drinks items")
    print("11. Display cart")
    print("12. Checkout")
    print("13. check balance")
    print("14. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        creat_bank_account()
    elif choice == "2":
        deposit_money()

    elif choice == "3":
        while True:
            name = input("Enter guest name: ")
            if name.replace(" ", "").isalpha():
                break
            else:
                print("Invalid name. Please enter a name with alphabets.")
        
        while True:
            email = input("Enter guest email: ")
            if "@" in email and email.endswith(".com"):
                break
            else:
                print("Invalid email. Please enter a correct email.")
        
        while True:
            phone_number = input("Enter guest phone number: ")
            if phone_number.isdigit() and len(phone_number) == 11:
                break
            else:
                print("Invalid phone number. Please enter a correct phone number.")
        
        while True:
            gender = input("Enter guest gender (male/female): ")
            if gender.lower() in ["male", "female"]:
                break
            else:
                print("Invalid gender. Please enter 'male' or 'female'.")
        
        while True:
            guest_origin = input("Enter guest origin: ")
            if guest_origin.isalpha():
                break
            else:
                print("Invalid origin. Please enter a correct origin.")
        
        while True:
            next_of_kin_name = input("Enter next of kin name: ")
            if next_of_kin_name.isalpha():
                break
            else:
                print("Invalid name. Please enter a name with alphabets.")
        
        while True:
            next_kin_gender = input("Enter next of kin gender (male/female): ")
            if next_kin_gender.lower() in ["male", "female"]:
                break
            else:
                print("Invalid gender. Please enter 'male' or 'female'.")
        
        while True:
            next_kin_phone_num = input("Enter next of kin phone number: ")
            if next_kin_phone_num.isdigit() and len(next_kin_phone_num) == 11:
                break
            else:
                print("Invalid phone number. Please enter a correct phone number.")
        
        guests[name] = {
            "email": email,
            "phone_number": phone_number,
            "gender": gender,
            "guest_origin": guest_origin,
            "next_of_kin_name": next_of_kin_name,
            "next_kin_gender": next_kin_gender,
            "next_kin_phone_num": next_kin_phone_num
        }
        print("Guest added successfully!")

    elif choice == "4":
        room_number = input("Enter room number: ")
        room_type = input("Enter room type (regular/special): ")
        
        if room_type.lower() == "regular":
            price = regular_room_price
        else:
            price = special_room_price
        
        rooms[room_number] = {
            "room_type": room_type,
            "price": price,
            "availability": "available"
        }
        print(f"Room {room_number} ({room_type}) added successfully with price ₦{price}!")

    elif choice == "5":
        while True:
            guest_name = input("Enter guest name: ")
            if name.replace(" ", "").isalpha():
                break
            else:
                print("Enter correct name")
        room_number = input("Enter room number: ")
        while True:
            check_in = input("Enter check-in date (YYYY-MM-DD): ")
            try:
                year, month, day = map(int, check_in.split('-'))
                if 1 <= month <= 12 and 1 <= day <= 31:
                    break
                else:
                    print("Invalid date. Please enter the date in YYYY-MM-DD format.")
            except:
                print("Invalid date format. Please enter the date in YYYY-MM-DD format.")            
        while True:
            check_out = input("Enter check-out date (YYYY-MM-DD): ")
            try:
                year, month, day = map(int, check_out.split('-'))
                if 1 <= month <= 12 and 1 <= day <= 31:
                    check_in_date = list(map(int, check_in.split('-')))
                    check_out_date = list(map(int, check_out.split('-')))
                    if check_out_date > check_in_date:
                        break
                    else:
                        print("Check-out date must be after check-in date.")
                else:
                    print("Invalid date. Please enter the date in YYYY-MM-DD format.")
            except:
                print("Invalid date format. Please enter the date in YYYY-MM-DD format.")

        if guest_name in guests and room_number in rooms and rooms[room_number]["availability"] == "available":
            bookings[room_number] = {
                "guest_name": guest_name,
                "check_in": check_in,
                "check_out": check_out,
                "room_price": rooms[room_number]["price"]
            }
            rooms[room_number]["availability"] = "occupied"
            print("Room booked successfully!")
        else:
            if guest_name not in guests:
                print("Guest not found.")
            elif room_number not in rooms:
                print("Room not found.")
            elif rooms[room_number]["availability"] != "available":
                print("Room not found or already occupied.")
        add_bookings_choice = input("Do you want to add this to your cart? (yes/no): ")
        if add_bookings_choice.lower() == "yes":
            while True:
                booking_date = input("Enter your booking date (YYYY-MM-DD): ")
                try:
                    year, month, day = map(int, booking_date.split('-'))
                    if 1 <= month <= 12 and 1 <= day <= 31:
                        if booking_date >= check_in and booking_date <= check_out:
                            break
                        else:
                            print("Booking date must be within the stay period.")
                    else:
                        print("Invalid date. Please enter the date in YYYY-MM-DD format.")
                except:
                    print("Invalid date format. Please enter the date in YYYY-MM-DD format.")

            while True:
                booking_time = input("Enter your booking time (HH:MM): ")
                try:
                    hour, minute = map(int, booking_time.split(':'))
                    if 0 <= hour <= 23 and 0 <= minute <= 59:
                        break
                    else:
                        print("Invalid time. Please enter the time in HH:MM format.")
                except:
                    print("Invalid time format. Please enter the time in HH:MM format.")
                    
                    

            if withdraw_money(guest_name, rooms[room_number]["price"]):
                add_to_cart("room", room_number, guest_name, 1, booking_date, booking_time)
                print(f"Booking added for {guest_name} on {booking_date} at {booking_time}!")  
            else:
                print("Insufficint funds.Cannot add bookig to cart.")
    elif choice == "6":
        room_number = input("Enter room number: ")
        if room_number in bookings:
            print(f"Guest {bookings[room_number]['guest_name']} checked in successfully!")
            del bookings[room_number]
        else:
            print("Room not booked.")
    elif choice == "7":
        room_number = input("Enter room number: ")
        if room_number in bookings:
            print(f"Guest {bookings[room_number]['guest_name']} checked out successfully!")
            del bookings[room_number]
            rooms[room_number]["availability"] = "available"
        else:
            print("Room not booked.")

    elif choice == "8":
        print("Food Menu:")
        for item_id, item in food_menu.items():
            print(f"{item_id}. {item['name']} - ₦{item['price']}")

        food_choice = input("Enter the number of your chosen food item: ")
        if food_choice in food_menu:
            item_id = food_choice
            item_name = food_menu[item_id]['name']
            item_price = food_menu[item_id]['price']
            print(f"You have chosen {item_name} for ₦{item_price}.")
            add_to_cart_choice = input("Do you want to add this to your cart? (yes/no): ")
            if add_to_cart_choice.lower() == "yes":
                guest_name = input("Enter your name: ")
                quantity = int(input("Enter quantity: "))
                if withdraw_money(guest_name,item_price * quantity):
                    add_to_cart("food", item_id, guest_name, quantity)
                    print("Food added to cart successfully!")
                else:
                    print("Insufficint funds.Cannot add bookig to cart.")

    elif choice == "9":
        print("Amenities:")
        for item_id, item in amenities.items():
            print(f"{item_id}. {item['name']} - ₦{item['price']}")

        amenity_choice = input("Enter the number of your chosen amenity: ")
        if amenity_choice in amenities:
            item_id = amenity_choice
            item_name = amenities[item_id]['name']
            item_price = amenities[item_id]['price']
            print(f"You have chosen {item_name} for ₦{item_price}.")
            add_to_cart_choice = input("Do you want to add this to your cart? (yes/no): ")
            if add_to_cart_choice.lower() == "yes":
                guest_name = input("Enter your name: ")
                if withdraw_money(guest_name,item_price ):
                    add_to_cart("amenity", item_id, guest_name)
                    print("Amenity added to cart successfully!")
                else:
                    print("Insufficint funds.Cannot add bookig to cart.")
                
                
    elif choice == "10":
        print("Drinks Menu:")
        for item_id, item in drinks_items.items():
            print(f"{item_id}. {item['name']} - ₦{item['price']}")

        drink_choice = input("Enter the number of your chosen drink: ")
        if drink_choice in drinks_items:
            item_id = drink_choice
            item_name = drinks_items[item_id]['name']
            item_price = drinks_items[item_id]['price']
            print(f"You have chosen {item_name} for ₦{item_price}.")
            add_to_cart_choice = input("Do you want to add this to your cart? (yes/no): ")
            if add_to_cart_choice.lower() == "yes":
                guest_name = input("Enter your name: ")
                quantity = int(input("Enter quantity: "))
                if withdraw_money(guest_name,item_price * quantity):
                    add_to_cart("drink", item_id, guest_name, quantity)
                    print("Drink added to cart successfully!")
                else:
                    print("Insufficint funds.Cannot add bookig to cart.")
                    

    

    elif choice == "11":
        guest_name = input("Enter your name: ")
        display_cart(guest_name)

    elif choice == "12":
        guest_name = input("Enter your name: ")
        if guest_name in cart:
            total_price = 0
            for item_type, items in cart[guest_name].items():
                for item in items:
                    total_price += item['price'] * item['quantity']
            print(f"Total price: ₦{total_price}")
            del cart[guest_name]
            print("Cart checked out successfully!")
        else:
            print("Cart is empty.")
    elif choice == "13":
        check_balance()

    elif choice == "14":
        print("Exixting  system, Have a great day.")
        break

    else:
        print("Invalid choice. Please try again.")

with open('hotel_managementss_data.csv', mode='a', newline='') as f:
    writer = csv.writer(f)
    # Write guests
    writer.writerow(['Guest Name', 'Email', 'Phone Number', 'Gender', 'Guest Origin', 'Next of Kin Name', 'Next of Kin Gender', 'Next of Kin Phone Number'])
    for guest_name, details in guests.items():
        writer.writerow([guest_name, details['email'], details['phone_number'], details['gender'], details['guest_origin'], details['next_of_kin_name'], details['next_kin_gender'], details['next_kin_phone_num']])
    
    
    
   # Write cart items
    writer.writerow(['Guest Name', 'Item Type', 'Item ID', 'Quantity', 'Booking Date', 'Booking Time'])
    for guest_name, items in cart.items():
        for item_type, item_list in items.items():
            for item in item_list:
                writer.writerow([guest_name, item_type, item['name'], item['quantity'], item.get('booking_date', ''), item.get('booking_time', '')])


    
```
